```python
# ============================================
# RQ7: Fraud Class Distribution Analysis
# ============================================

import pandas as pd
import matplotlib.pyplot as plt

# --------------------------------------------
# Step 1: Load Dataset
# --------------------------------------------

df = pd.read_csv(
    r"C:\Users\dheer\Desktop\ML\cleaned_dataset.csv"
)

print("Dataset Loaded Successfully")

# --------------------------------------------
# Step 2: Define Target Column
# --------------------------------------------

target_col = 'Is_Fraudulent'

# --------------------------------------------
# Step 3: Calculate Class Distribution
# --------------------------------------------

class_counts = df[target_col].value_counts()

labels = ['Non-Fraudulent', 'Fraudulent']

sizes = [
    class_counts.get(0, 0),
    class_counts.get(1, 0)
]

print("\nClass Distribution:")
print(class_counts)

# --------------------------------------------
# Step 4: Create Donut Chart
# --------------------------------------------

plt.figure(figsize=(8, 8))

wedges, texts, autotexts = plt.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%',
    startangle=90,
    wedgeprops=dict(width=0.4)
)

# Center Circle for Donut Shape
centre_circle = plt.Circle((0, 0), 0.70, fc='white')

fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title('RQ7: Fraud vs Non-Fraud Claim Distribution')

# --------------------------------------------
# Step 5: Save Results
# --------------------------------------------

plt.savefig(
    r"C:\Users\dheer\Desktop\ML\RQ7_donut_chart.png",
    bbox_inches='tight'
)

class_counts.to_csv(
    r"C:\Users\dheer\Desktop\ML\RQ7_class_distribution.csv"
)

plt.show()

print("\nProcess Completed Successfully")
```
