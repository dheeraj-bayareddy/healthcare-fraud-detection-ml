```python
# ======================================
# PREPROCESSING NOTEBOOK (CLEAN VERSION)
# ======================================

import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder

# --------------------------------------
# Step 1: Load Dataset
# --------------------------------------

df = pd.read_csv(
    r"C:\Users\dheer\Downloads\Synthetic_health_insurance_claims_fraud_detection.csv"
)

print("✅ Dataset Loaded")
print("Original Shape:", df.shape)

# --------------------------------------
# Step 2: Remove Missing Values
# --------------------------------------

df = df.dropna()

print("After Removing Nulls:", df.shape)

# --------------------------------------
# Step 3: Reduce Dataset Size
# --------------------------------------

df = df.sample(n=50000, random_state=42)

print("After Sampling:", df.shape)

# --------------------------------------
# Step 4: Detect Fraud Column
# --------------------------------------

target_col = None

for col in df.columns:
    if 'fraud' in col.lower():
        target_col = col
        break

print("Detected Target Column:", target_col)

# --------------------------------------
# Step 5: Encode Categorical Columns
# --------------------------------------

le = LabelEncoder()

categorical_cols = df.select_dtypes(
    include=['object', 'string']
).columns

for col in categorical_cols:
    df[col] = le.fit_transform(df[col].astype(str))

# --------------------------------------
# Step 6: Convert Boolean Target
# --------------------------------------

if df[target_col].dtype == 'bool':
    df[target_col] = df[target_col].astype(int)

print("Final Dataset Shape:", df.shape)

# --------------------------------------
# Step 7: Save Cleaned Dataset
# --------------------------------------

save_path = r"C:\Users\dheer\Desktop\cleaned_dataset.csv"

df.to_csv(save_path, index=False)

print("✅ Cleaned dataset saved successfully")
print("Saved at:", save_path)
```
